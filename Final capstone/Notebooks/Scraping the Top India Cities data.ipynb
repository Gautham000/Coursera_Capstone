#### Import all the required packages

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

#### Import the wikipedia page

```python
url="https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
```

```python
res=requests.get(url).text
soup = BeautifulSoup(res, 'lxml')
print(soup.prettify)
```

#### Scrape the wikipedia page to get the required data

```python
table = soup.find('table', {'class': 'wikitable sortable'})
rows = table.find_all("tr")
data = []
for tr in rows:
    td = tr.find_all("td")
    row = [tr.text for tr in td]
    data.append(row)
data.pop(0)
data
```

```python
cols=['index1','City', 'Population','Population2001','State']
df=pd.DataFrame(columns=cols, data=data)
df.head()
```

```python
df.drop(['index1','Population2001'],axis=1, inplace=True)
```

```python
df['State'] = df['State'].apply(lambda x: x.rstrip('\n'))
df
```

```python
df['City'] = df['City'].apply(lambda x: x.rstrip('[*'))
df
```

```python
df.set_value(16,'City','Visakhapatnam')
```

save the data as dataset

```python
df.to_csv("India top cities.csv")
```
