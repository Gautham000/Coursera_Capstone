# Geospatial location calcution

Geospatial locations are calculated with the help of geocoder.

```python
def location(area):
    address= area+', IN'
    geolocator = Nominatim(user_agent = 'my-application')
    location = geolocator.geocode(address)
    lat = location.latitude
    lon = location.longitude
    latitude.append(lat)
    longitude.append(lon)
```

```python
latitude=[]
longitude=[]

for i in data['City']:
    location(i)
print(latitude,longitude)
```

```python
location=pd.DataFrame({'Latitude':latitude, 'Longitude':longitude})
location.head()
```

```python
frames=[data,location]
City_location=data.merge(location, left_index=True, right_index=True, how='outer')
```
